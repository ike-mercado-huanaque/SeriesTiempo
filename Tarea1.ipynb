```r
siniestrados <- read_excel('/Users/IkeMercado/Desktop/USM/2020-1/Series Tiempo/Siniestrados.xlsx')
head(siniestrados)
```

```r
deaths <- read.table('/Users/IkeMercado/Desktop/USM/2020-1/Series Tiempo/deaths.txt')
```

```r
head(deaths)
```

```r
matplot(x=deaths$V2,
        y=deaths$V3,
        main="Muertes por mes en EEUU",
        type="l",
        xlab="Años",
        ylab="Muertes",
        col=c("red"),
    );
```

```r
matplot(x=siniestrados$Año,
        y=cbind(siniestrados$Lesionados_Graves,
                siniestrados$Lesionados_Menos_graves,
                siniestrados$Lesionados_Leves),
        xlab="Años",
        ylab="Lesionados",
        main="Numero de lesionados por año",
        type="l",
        col=cbind("blue","red","green"),
       )
legend(x="topleft",legend=cbind("Graves","Menos graves","Leves"),fill=c("blue","red","green"))
```

```r
matplot(x=siniestrados$Año,
        y=cbind(siniestrados$Siniestros),
        xlab="Años",
        ylab="Cantidad de personas",
        main="Numero de Siniestrados por año",
        type="l",
        col=cbind("blue","red"),
       )
```

```r
matplot(x=siniestrados$Año,
        y=cbind(siniestrados$Fallecidos),
        xlab="Años",
        ylab="Cantidad de personas",
        main="Numero de Fallecidos por año",
        type="l",
        col=cbind("blue","red"),
       )
```

```r
matplot(x=siniestrados$Año,
        y=cbind(siniestrados$Total_lesionados,
                siniestrados$Total_víctimas),
        xlab="Años",
        ylab="Cantidad de personas",
        main="Numero de lesionados por año",
        type="l",
        col=cbind("blue","red"),
       )
legend(x="topleft",
       legend=cbind("Total Lesionados",
                                "Total Víctimas"),
       fill=c("blue","red"))
```

```r
matplot(x=siniestrados$Año,
        y=cbind(siniestrados$Parque_vehicular),
        xlab="Años",
        ylab="Parque vehicular",
        main="Parque vehicular por año",
        type="l",
        col=cbind("blue","red"),
       )
```

```r
matplot(x=siniestrados$Año,
        y=cbind(siniestrados$Población),
        xlab="Años",
        ylab="Cantidad de personas",
        main="Poblacion por año",
        type="l",
        col=cbind("red"),
       )
```

```r
matplot(x=siniestrados$Año,
        y=cbind(siniestrados$Indicadores_cada_10000_vehículos_Mortalidad,
                siniestrados$Indicadores_cada_10000_vehículos_Siniestralidad),
        xlab="Años",
        ylab="Cantidad de personas",
        main="Mortalidad y siniestralidad por año por cada 10000 vehiculos",
        type="l",
        col=cbind("blue","red"),
       )
legend(x="topleft",
       legend=cbind("Mortalidad cada 10000 vehiculos",
                    "Siniestralidad cada 10000 vehiculos"),
       fill=c("blue","red"))
```

```r
matplot(x=siniestrados$Año,
        y=cbind(siniestrados$Indicadores_cada_100000_habitantes_Mortalidad,
                siniestrados$Indicadores_cada_100000_habitantes_Siniestralidad),
        xlab="Años",
        ylab="Cantidad de personas",
        main="Mortalidad y siniestralidad por año por cada 100000 Habitantes",
        type="l",
        col=cbind("blue","red"),
       )
legend(x="topleft",
       legend=cbind("Mortalidad cada 100000 hab",
                    "Siniestralidad cada 100000 hab"),
       fill=c("blue","red"))
```
