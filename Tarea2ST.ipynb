```r
mean = 0;
var = 1;
n = 100;
x <- rnorm(n,mean,var);
```

```r
#AR(1)
#Serie1
phi = 0.25
x0 <- 3
AR11 <- numeric(n)
z <- rnorm(n,mean,var)
AR11[1] = x0
for(i in 2:n){
    AR11[i] <- z[i]+phi*AR11[i-1]
}
```

```r
#Serie2
phi = 0.80
x0 <- -10
AR12 <- numeric(n)
z <- rnorm(n,mean,var)
AR12[1] = x0
for(i in 2:n){
    AR12[i] <- z[i]+phi*AR12[i-1]
}
```

```r
#AR(2)
#Serie 1
phi = c(0.5,0.4)
x0 <- 3
x1 <- 2
AR21 <- numeric(n)
z <- rnorm(n,mean,var)
AR21[1:2] = c(x0,x1);
for(i in 3:n){
    AR21[i] <- z[i]+phi[1]*AR21[i-1]+phi[2]*AR21[i-2]
}

#AR(2)
#Serie 2
phi = c(0.1,0.8)
x0 <- 5
x1 <- -3
AR22 <- numeric(n)
z <- rnorm(n,mean,var)
AR22[1:2] = c(x0,x1);
for(i in 3:n){
    AR22[i] <- z[i]+phi[1]*AR22[i-1]+phi[2]*AR22[i-2]
}
```

```r
#AR(3)
#Serie 1
phi = c(0.2,0.4,0.3)
x0 <- 3
x1 <- 2
x2 <- 2.5
AR31 <- numeric(n)
z <- rnorm(n,mean,var)
AR31[1:3] = c(x0,x1,x2);
for(i in 4:n){
    AR31[i] <- z[i]+phi[1]*AR31[i-1]+phi[2]*AR31[i-2]+phi[3]*AR31[i-3]
}

#AR(3)
#Serie 2
phi = c(0.25,0.01,0.5)
x0 <- 5
x1 <- -3
AR32 <- numeric(n)
z <- rnorm(n,mean,var)
AR32[1:3] = c(x0,x1,x2);
for(i in 4:n){
    AR32[i] <- z[i]+phi[1]*AR32[i-1]+phi[2]*AR32[i-2]+phi[3]*AR32[i-3]
    }
```

```r
#MA(1)
#Serie1
phi = 0.25
z <- rnorm(n,mean,var)
MA11 <- numeric(n)
MA11[1:2] <- z[1:2]
for(i in 3:n){
    MA11[i] <- z[i]+phi*z[i-1]
}
#Serie2
phi = 0.80
z <- rnorm(n,mean,var)
MA12 <- matrix(0,1,n)
MA12[1:2] <- z[1:2]
for(i in 3:n){
    MA12[i] <- z[i]+phi*z[i-1]
}
```

```r
#MA(1)
#Serie1
phi = 0.25
z <- rnorm(n,mean,var)
MA11 <- numeric(n)
MA11[1:2] <- z[1:2]
for(i in 2:n){
    MA11[i] <- z[i]+phi*z[i-1]
}
#Serie2
phi = 0.80
z <- rnorm(n,mean,var)
MA12 <- numeric(n)
MA12[1:2] <- z[1:2]
for(i in 2:n){
    MA12[i] <- z[i]+phi*z[i-1]
}
```

```r
#MA(2)
#Serie1
phi = c(0.5,0.4)
z <- rnorm(n,mean,var)
MA21 <- numeric(n)
MA21[1:3] <- z[1:3]
for(i in 3:n){
    MA21[i] <- z[i]+phi[1]*z[i-1]+phi[2]*z[i-2]
}
#Serie2
phi = c(0.1,0.8)
z <- rnorm(n,mean,var)
MA22 <- numeric(n)
MA22[1:3] <- z[1:3]
for(i in 3:n){
    MA22[i] <- z[i]+phi[1]*z[i-1]+phi[2]*z[i-2]
}
```

```r
#MA(3)
#Serie1
phi = c(0.2,0.4,0.3)
z <- rnorm(n,mean,var)
MA31 <- numeric(n)
MA31[1:4] <- z[1:4]
for(i in 4:n){
    MA31[i] <- z[i]+phi[1]*z[i-1]+phi[2]*z[i-2]+phi[3]*z[i-3]
}
#Serie2
phi = c(0.25,0.01,0.5)
z <- rnorm(n,mean,var)
MA32 <- numeric(n)
MA32[1:4] <- z[1:4]
for(i in 4:n){
    MA32[i] <- z[i]+phi[1]*z[i-1]+phi[2]*z[i-2]+phi[3]*z[i-3]
}
```
